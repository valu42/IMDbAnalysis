```python
import pandas as pd
import numpy as np
```

```python
rating_data_fn = "rating_data.tsv"
basic_data_fn = "basic_data.tsv"
merged_fn = "merged.csv"
```

```python
basic_dtypes = {"tconst": str, 
                "titleType": str, 
                "primaryTitle": str, 
                "originalTitle": str,
                "isAdult": str,
                "startYear": str,
                "endYear": str,
                "runtimeMinutes": str,
                "genres": str,
               }
ratings_dtypes = {
    "tconst": str,
    "averageRating": str,
    "numVotes": int,
}

basics_df = pd.read_csv(basic_data_fn, sep="\t", low_memory=False, dtype=basic_dtypes)
ratings_df = pd.read_csv(rating_data_fn, sep="\t", low_memory=False, dtype=ratings_dtypes)
```

```python
merged = pd.merge(basics_df, ratings_df)
```

```python
merged
```

```python
merged.to_csv(merged_fn, index=False)
```
